```python
num_cycle=2
min_lr=5
max_lr=15
min_to_max=list(range(min_lr, max_lr))       
max_to_min=list(range(max_lr, min_lr-1,-1))  
final_list=[]
final_list.extend(min_to_max)
final_list.extend(max_to_min)

for i in range(num_cycle) :
  final_list.extend(final_list)
  
```

```python
import matplotlib.pyplot as plt
plt.plot(final_list)
plt.title('Triangular schedule')
plt.show()
```
